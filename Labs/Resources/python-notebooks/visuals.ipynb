```python
import pandas as pd
# Sample data
data = {
    "channel": ["Online", "In-Store", "Online", "In-Store", "Mobile"],
    "revenue": [200, 350, 150, 400, 180],
    "transactions": [50, 80, 40, 90, 45]
}
 
df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(data=df, x="channel", y="revenue")
plt.title("Revenue by Channel")
plt.show()
```

```python
sns.scatterplot(data = df, x = "transactions", y = "revenue", hue = "channel")
plt.show()
```
